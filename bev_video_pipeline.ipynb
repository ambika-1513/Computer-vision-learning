```python
import torch
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
torch.cuda.get_device_name(0)
```

```python
cap  = cv2.VideoCapture("traffic.mp4")
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
```

```python
out_writer = cv2.VideoWriter("output_bev.mp4", cv2.VideoWriter_fourcc(*"mp4v"), fps, (1100 + width, 720))
```

```python
frame_num = 0
```

```python
src_points = np.float32([
    [150, 650],
    [1000, 650],
    [650, 150],
    [480, 150]
])
dst_points = np.float32([
    [200, 700],
    [900, 700],
    [900, 0],
    [200, 0]
])

matrix = cv2.getPerspectiveTransform(src_points, dst_points)
```

```python
while True:
  ret, frame = cap.read()
  if not ret:
    break
  frame_num += 1
  frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
  bev = cv2.warpPerspective(frame_rgb, matrix, (1100, 720))
  combined = np.hstack((frame_rgb, bev))
  out_writer.write(combined)
  if frame_num % 30 == 0:
    print(f"Frame {frame_num} processed")
cap.release()
out_writer.release()
```
